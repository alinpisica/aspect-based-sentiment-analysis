```python
import pandas
```

```python
MODELS = [
    ('bert_pre_trained_dropout_linear_512', 'BERT PT+Dropout+Linear'),     
    ('bert_pre_trained_dropout_bilstm_linear_512', 'BERT PT+Dropout+BiLSTM+Linear'),
    ('bert_pre_trained_dropout_cnn_bilstm_linear_512', 'BERT PT+Dropout+CNN+BiLSTM+Linear'),
    ('bert_fine_tuned_dropout_linear_512', 'BERT FT+Dropout+Linear'),
    ('bert_fine_tuned_dropout_bilstm_linear_512', 'BERT FT+Dropout+BiLSTM+Linear'),
    ('bert_fine_tuned_dropout_cnn_bilstm_linear_512', 'BERT FT+Dropout+CNN+BiLSTM+Linear'),
]
```

```python
PATHS = [
    ('./results/ATE/SemEval16 - Task 5 - Restaurants/stats/', 'ATE SemEval16 - Task 5 - Restaurants (BERT)'),
    ('./results/ATE/MAMS/stats/', 'ATE MAMS (BERT)'),
    ('./results/ABSA/SemEval16 - Task 5 - Restaurants/stats/', 'ABSA SemEval16 - Task 5 - Restaurants (BERT)'),
    ('./results/ABSA/MAMS/stats/', 'ABSA MAMS (BERT)'),
]
```

```python
for path, name in PATHS:
    print(name)
    results = pandas.DataFrame(columns=['Model', 'Accuracy', 'Precision (micro)', 'Precision (macro)', 'Recall (micro)', 'Recall (macro)', 'F1 (micro)', 'F1 (macro)', 'Execution time (s)'])

    for model, model_name in MODELS:
        df = pandas.read_csv(path + model + '.csv')

        print('\t' + model_name)

        accuracy = "{:.2f}".format(df['accuracy'].mean() * 100)
        precision_micro = "{:.2f}".format(df['precision_score_micro'].mean() * 100)
        precision_macro = "{:.2f}".format(df['precision_score_macro'].mean() * 100)
        recall_micro = "{:.2f}".format(df['recall_score_micro'].mean() * 100)
        recall_macro = "{:.2f}".format(df['recall_score_macro'].mean() * 100)
        f1_micro = "{:.2f}".format(df['f1_score_micro'].mean() * 100)
        f1_macro = "{:.2f}".format(df['f1_score_macro'].mean() * 100)
        execution_time = "{:.2f}".format(df['execution_time'].mean())
        
        print('\t\tAccuracy: ' + accuracy + '%')
        print('\t\tPrecision (micro): ' + precision_micro + '%')
        print('\t\tPrecision (macro): ' + precision_macro + '%')
        print('\t\tRecall (micro): ' + recall_micro + '%')
        print('\t\tRecall (macro): ' + recall_macro + '%')
        print('\t\tF1 (micro): ' + f1_micro + '%')
        print('\t\tF1 (macro): ' + f1_macro + '%')
        print('\t\tExecution time: ' + execution_time + 's')

        results = results.append({'Model': model_name, 'Accuracy': accuracy, 'Precision (micro)': precision_micro, 'Precision (macro)': precision_macro, 'Recall (micro)': recall_micro, 'Recall (macro)': recall_macro, 'F1 (micro)': f1_micro, 'F1 (macro)': f1_macro, 'Execution time (s)': execution_time}, ignore_index=True)
    
    results.to_csv(path + 'results.csv', index=False)

    print('\n')
```
